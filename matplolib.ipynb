```python
Создать список из 6 случайных целых чисел в диапазоне от 2 до 10 (включительно). 
По ним построить Pie chart. 
Поменять базовые цвета на какие-нибудь монохромные. 
Добавить подписи относительных пропорций каждой величины относительно общего. 
Отделить один кусочек от остальных.
```

```python
import matplotlib.pyplot as plt
```

```python
import random
y = []
for i in range (6):
    a = random.randrange(2,11,1)
    y.append(a)
fig = plt.figure()
explode = (0.0, 0.0, 0.0, 0.0, 0.2, 0.0)
colors = ( "#B0080F", "#B0020F", "#B0198F", 
          "#B0362F", "#B0260F", "#B0155F") 
plt.pie (y, autopct = '%1.2f%%', explode = explode, colors = colors)
plt.show()
```
